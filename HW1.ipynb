```python
import re

cut_words=""
for line in open('test.txt',encoding='utf-8'):
    line.strip('\n')
    line = re.sub("[A-Za-z0-9\：\·\—\，\。\“ \”\.]", "", line)
    seg_list=jieba.cut(line,cut_all=False)
    cut_words+=(" ".join(seg_list))
all_words=cut_words.split()

clear_list = []
for seg in all_words:
    if len(seg) > 1:
        clear_list.append(seg)
        
print(len(clear_list))
```

```python
import pandas as pd

def count_segment_freq(seg_list):
  seg_df = pd.DataFrame(seg_list,columns=['seg'])
  seg_df['count'] = 1
  sef_freq = seg_df.groupby('seg')['count'].sum().sort_values(ascending=False)
  sef_freq = pd.DataFrame(sef_freq)
  return sef_freq

sef_freq = count_segment_freq(clear_list)
#sef_freq.head()
sef_freq
```

```python
from collections import Counter
import pandas as pd

c=Counter()
for x in clear_list:
    if len(x)>1 and x != '\r\n':
        c[x] += 1

key_list = []
freq_list = []
tmp = []
tmp1 = []
tmp2 =[]
print('\n結果：')
count = 0
for (k,v) in c.most_common(100):
    key_list.append(k)
    freq_list.append(v)
    
for (k,v) in c.most_common(len(c)):
    tmp1.append(k)
    count+=1
    tmp2.append(count)
pairs = pd.Series(tmp2, index=tmp1)

for key in key_list:
    tmp.append(pairs[key])
count = pd.Series(freq_list, index=tmp)
#count
count.plot()
```
