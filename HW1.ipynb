```python
import requests
import json
response = requests.get(
    "https://raw.githubusercontent.com/cjwu/cjwu.github.io/master/courses/nlp/hw1-dataset.txt")
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")
  #輸出排版後的HTML內容
file = open(r'C:\NLP_HW1\test.txt', 'w', encoding = 'UTF-8')
file.write(soup.prettify())
file.close()
```

```python
import jieba 
import jieba.analyse
import matplotlib.pyplot as plt
import pandas as pd
word_lst = []
weight_lst=[]
#for line in open('test.txt', encoding = 'utf8'):#1.txt是需要分詞統計的文件
#    item = line.strip('\n\r').split('\t') #製表格切分
        # print item
#    tags = jieba.analyse.extract_tags(item[0]) #jieba分詞
#    for t in tags:
#        word_lst.append(t)
#word_lst
file = open('test.txt','r', encoding = 'utf8')
txt = file.read()
for tags, weight in jieba.analyse.extract_tags(txt, topK=100, withWeight=True):
    word_lst.append(tags)
    weight_lst.append(weight)
#topK 為返回幾個TF/IDF 權重最大的關鍵詞，默認值為20
word_lst[0]
#tags = jieba.analyse.extract_tags(txt, topK=2)  
#for line in open('test.txt'):
#tags   
#with open("")
```

```python
plt.figure(figsize=(30,10))
plt.rcParams['font.sans-serif'] = ['Taipei Sans TC Beta'] 
plt.grid()
plt.plot(word_lst,weight_lst)
```

```python
count = []
for item in word_lst:
    #print('occurrence of letter :',item, txt.count(item))
    count.append(txt.count(item))
    
```
